```python
topics_string = """
Zebra Crossings,Buildings,Bridges,Roads,Traffic
"""
NUM_DOWNLOADS = 500
def clean_string_convert(x):
    lst = x.split(",")
    
    new_lst = []
    for each in lst:
        each = each.strip()
        if not each=="":
            new_lst.append(each.title())
    return new_lst

topics = clean_string_convert(topics_string)
topics
```

```python
# the json file where the output must be stored 
import json
out_file = open("data_file.json", "w") 
    
json.dump({"topics":topics}, out_file, indent = 6) 
```

```python
from bing_image_downloader import downloader
out = 'dataset'
for topic in topics:
    downloader.download(topic.title(), limit=NUM_DOWNLOADS,  output_dir=out, adult_filter_off=True, force_replace=False, timeout=60)
```

```python
import shutil
def zip_and_remove(x):
    shutil.make_archive(x, 'zip',  root_dir=x)
    shutil.rmtree(x)
zip_and_remove(out)
```
